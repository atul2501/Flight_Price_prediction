```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.drop(columns=['Total_Stops']),
                                                          df['Total_Stops'],
                                                          test_size=0.2,
                                                          random_state=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#one hot Encoder
trf1=ColumnTransformer(
    [
        ('ohe_Air_source_des',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[0,1,2,3])
    ],
remainder='passthrough')
```

```python
X_train
```

```python
X_train['Destination'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
#scaled for better performace
trf3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,40))
])
```

```python
# from sklearn.feature.selection import SelectKBest,chi2
from sklearn.feature_selection import SelectKBest,chi2
#Feature selection 
trf4=SelectKBest(score_func=chi2,k=8)
```

```python
from sklearn.tree import DecisionTreeClassifier
trf5=DecisionTreeClassifier()
```

```python
from sklearn.pipeline import Pipeline
#Create Pipeline
pipe=Pipeline([
    ('trf1',trf1),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
    
])
```

```python

pipe.fit(X_train,y_train)
```

```python
#Predict
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Cross Validation using pipeline
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=100,scoring='accuracy').mean()
```

```python

#GridSerch using Pipeline
params={
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(pipe,params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python

grid.best_score_
```
